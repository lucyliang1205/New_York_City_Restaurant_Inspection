2.Output a DataFrame that consists of the DBA, BORO, GRADE and SCORE columns, sorted by BORO. Remove rows that contain null values and rows with a value of 0 in the BORO column. 5pts

```python
#sorted BORO
dataframe=df_restaurant[["DBA","BORO","GRADE","SCORE",]].dropna().sort_values("BORO")
#remove row with 0 
dataframe= dataframe[dataframe["BORO"] != '0']
#output data
dataframe
```

3.Print the total number of rows that have the following strings as part of the values in the DBA column:
DUNKIN

POPEYES

MCDONALDS

STARBUCKS

```python
#count the total number of each DBA
dunkinCount = len(dataframe[dataframe["DBA"].str.contains("DUNKIN")])

popeyesCount = len(dataframe[dataframe["DBA"].str.contains("POPEYES")])

mcdonaldsCount = len(dataframe[dataframe["DBA"].str.contains("MCDONALDS")])

starbucksCount = len(dataframe[dataframe["DBA"].str.contains("STARBUCKS")])

print("DUNKIN : {}".format(dunkinCount))
print("POPEYES :{}".format(popeyesCount))
print("MCDONALDS : {}".format(mcdonaldsCount))
print("STARBUCKS : {}".format(starbucksCount))
```

4.Using the DataFrame created in question 2, create and output a Series that contains the mean SCORE for each BORO. Use the borough name for the index of the Series.

```python
boro=dataframe.groupby('BORO')['SCORE'].mean()
boro
```

5.Accept user input of a zip code. Based on that zip code, create a pie chart that plots the percentage of each letter grade that makes up the total. If the user does not enter a zip code that exists in the csv file, print an error message accordingly. The title of the graph should be the zip code provided by the user and a legend should be included. 

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
zipAsk=int(input("Which zip code are you looking for the grade?"))
if zipAsk in df_restaurant['ZIPCODE'].unique():
    grade = df_restaurant[df_restaurant['ZIPCODE'] == zipAsk]['GRADE'].value_counts()
    plt.figure(figsize=(6,6))
    plt.title(zipAsk, fontsize=14)
    plt.pie(grade, autopct='%1.1f%%', labels=grade.index)
    plt.legend()
    plt.show()
else:
    print("Error message!")
```
